```python
from lxml import html  
import requests
from time import sleep


from collections import OrderedDict
from time import sleep

import os
import json
import traceback
import pandas as pd
import numpy as np
```

```python
def JSONParser(tickers):
    
    new_array_dataframes = pd.DataFrame()
    i = 0
    for ticker in tickers:
        url_price = "https://query1.finance.yahoo.com/v10/finance/quoteSummary/%s?modules=price"%(ticker)
        url_summaryprofile = "https://query1.finance.yahoo.com/v10/finance/quoteSummary/%s?modules=summaryProfile"%(ticker)
        print("Parsing Price URL: %s"%(url_price))
        print("Parsing Profile URL: %s"%(url_price))

        #Handling the exception for request and response
        try:
            price_response_object = requests.get(url_price, headers={'User-agent': 'Mozilla/5.0'})
            profile_response_object = requests.get(url_summaryprofile, headers={'User-agent': 'Mozilla/5.0'})
            json_loaded_price =  json.loads(price_response_object.text)
            json_loaded_summary =  json.loads(profile_response_object.text)
            print(json_loaded_price)
            print(json_loaded_summary)
        except:
            print('Error Occured while executing the request')
            traceback.print_exc()
            return {"error":"Failed to parse json response"}


        price_keys = json_loaded_price['quoteSummary']['result'][0]['price']
        profile_keys = json_loaded_summary['quoteSummary']['result'][0]['summaryProfile']
        print((price_keys))
        print(profile_keys)

        # create the dictionary from the JSON object
        dict_prices = {}
        for k,v in price_keys.items():
            if type(v) == dict:
                if len(v) == 0:
                    dict_prices[k] = 'NAN'
                else:
                    dict_prices[k] = v.get("fmt")
            else:
                if (v) == None:
                    dict_prices[k] = 'NAN'
                else:
                    dict_prices[k] = v

        df_prices = pd.DataFrame(data=dict_prices, index = [i])

        new_array_dataframes = pd.concat([new_array_dataframes, df_prices], axis = 0)
        i += 1
  
    print((new_array_dataframes))
    
    strPathDir = os.getcwd()
    strPathDir += "/Desktop/output.xlsx"

    new_array_dataframes.to_excel(strPathDir, sheet_name="prices")
  


```

```python
tickers = ['AAPL', 'AAP', 'GOOG']
JSONParser(tickers)

    



```
